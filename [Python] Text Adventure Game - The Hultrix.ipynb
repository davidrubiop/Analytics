```python
##### October 2019
##### Text Adventure Game - Data Science: Python
##### by David Rubio - MSc Business Analytics
##### Hult International Business School

"""
DocString:
    1) 
    Intro:
    This is 'THE HULTRIX', a small tribute to 'The Matrix',
    one of the movies that, as a boy, aroused my interest 
    for computers and technology in general.
    Also, after almost 20 years, 'Matrix 4' will be released in 2021!

    Stages:
    Start
    1. The Pills: Decide to take 1 of 2 pills.
    2. The Interaction: Select a place for your training.
    3. The Training: 2 chances of guessing 1 characteristic of the enemy. 
    4. The Fight: multiple attacks to the enemy while suffering the consequences.
    (Win / Lose) -> Start Again

    References:
    If you did not watch the movie, check this video with the scene
    in which stage 1 of this game is based: https://www.youtube.com/watch?v=zE7PKRjrid4
    Also, check its IMDB Page: https://www.imdb.com/title/tt0133093/

    2) 
    Known Bugs and/or Errors:
    None.
"""

###Import statements for system, time and random
import sys
from random import randint
import time

### Defining Animation Function
def context (message):
    """This function creates the effect of animation in the text while
    one of the characters is saying something.
    This will be used instead of 'print()' several times"""
    
    global context     
    for letter in message: #For Loop with iterator 'letter'
        print(letter, end = '', flush = True)
        time.sleep(0.015) #This will be the delay per letter (in seconds) 
        
### Defining Starting Function       
def start_game():
    """This function is the trigger that 'executes' the game.
    Introduces the user and the guide, who quickly asks for name and age to customize the experience.
    Several variables are defined.
    Used nested conditionals within a loop to make fun age validations (easter eggs)"""
    
    #Declaring Global Variables
    global guide_name
    global user_name
    global user_age
    global user_place
    
    guide_name = '* Mysterious Man' #Defined a generic name to keep the guide's real name anonymous
    print('_'* 50)
    print('_'* 40)
    print('_'* 30)
    
    #ASCII IMAGE FOR PRESENTATION PURPOSES
    print("""
                        WELCOME TO THE HULTRIX
    
    
                          ..sSs$$$$$$b.                                       
                       .$$$$$$$$$$$$$$$.                                     
                    .$$$$$$$$$$$$$$$$$$$$$b.                                 
                  .$$$$$$$$$$$$$$$$$$$$$$$$$                                 
                 $$$$$$$$$$$$$$$$$S'   `$$$$                                 
                 $$$$$$$$$$$$$$S'        $$$                                 
                 $$$$$$$$$$$$$'          `$$.                                
                 `$$$$$$$$$$$$$           `$$$.                              
                   `$$$$$$$$$'       .s$$$ $$ $                              
                     $$$$$$$$$.sSs .s$$s"   s s                               
                      $$$$$$$$$$$$           $P                              
                      `$$$$$$$$$$$s          $                               
                        $$$$$$$$$$$.    ",                                   
                        `$$$$$$$$$$sS$                                       
        s$$$.            `$$$$$$$$$$$$.s"'   .$.                              
        $$$$$.              `$$$$$$$$$$.    .$$$Ss.s$s.                      
         $$$$$.              `$$$$$$$$$P   .$$$$$$$$$$$$.                    
         $$$$$$.               `$$$$$$$' .$$$$$$$$$$$$$$$$.                  
         `$$$$$$.                 $$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$s.         
           $$$$$$.                `$$$$$$$$$ $$$$$$$$$$$$$$$$$$$$$$$$s.      
         .s$$$$$$$.                 `$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$.    
         s  $$$$$$$.                .$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$.   
         `$$$$$$$$$$.             .$$$' $$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$. 
          `$$$$$$$$$$.           s$'   $$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$. 
          $$$$$$$$$$$$e         $$$     `$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$ 
        .' $$$$$$$$$$7         $$$$       `$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
       "   `$$$$$$$$7         $$$$$       .$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
      $$Ss..$$$$$$$7        $$$$$$$    .s$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
       $$$$$$$$$$$$        $$$$$$$$ .s$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
        $$$$$$$$$$$     .$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
      .$$$$$$$$$$$$$   .$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
      $$$$$$$$$$$$$$  .$$$$$$$$$$$ $$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
          """)
    print('_'* 30)
    print('_'* 40)
    print('_'* 50)
    
    prompt("< Press Enter to start playing > \n") #These inputs will work as flow controls throughout the game        
    
    ###First time calling function 'context()' to replace 'print()'
    context("""
 Context: 
 < You just opened your eyes. It's all dark.
 You don't know where you are.
 Suddenly, a mysterious man gets into the room.>
    """)
    input("< Press Enter to continue > \n")
    
    context(f"""
 {guide_name}: Hello there! You finally woke up.
 What's your name?
        """)
    
    #User defines his/her name
    name = input("< How would you like to be called? >\n ")
    user_name = name.capitalize()
    time.sleep(1)
    context(f"{guide_name}: {user_name}??? What a coincidence!\n")
    context(f"""      
 {user_name}: What do you mean?\n
 {guide_name}: Nevermind, I'm just curious ... How old are you {user_name} ?\n
 {user_name}: Why do you need to know that? Who are you?\n
        """)
    input("< Press Enter to continue > \n")
    print("_"*50)
    
    #Mystery Man reveals his name and more info   
    context(f"""
 {guide_name}: My name is Morpheus. I'm the leader of the resistance.\n 
 Now, please tell me your age, we don't have much time!\n
         """) 
    guide_name = '* Morpheus'
    age = input("<Insert here your age using integer numbers i.e 30 >\n")
    
    #Age validation
    while type(age) != int:
        try:
            user_age = int(age) 
            if user_age >=5 and user_age <= 85: #Decided to be strict with the age. Plus a couple of easter eggs
                if user_age < 17:
                    print("""
                 ,=""=,
               c , _,{
               /\  @ )                 __
              /  ^~~^\          <=.,__/ '}=
             (_/ ,, ,,)          \_ _>_/~
              ~\_(/-\)'-,_,_,_,-'(_)-(_)  
 This game is rated as 'Mature' by the Entertainment Software 
 Rating Board (ESRB) which means you have to be 17+ years old 
 to be able to play ;) .\n

                    """)
                    endgame() #This takes us to the Fail Scenario
                    break
            
                elif user_age > 85:
                    print(f"""
                                  
                    ,:' `..;
                    `. ,;;'%
                    +;;'%%%%%
                     /- %,)%%
                     `   \ %%
                      =  )/ \
                      `-'/ / \
                        /\/.-.\
                       |  (    |
                       |  |   ||
                       |  |   ||
                   _.-----'   ||
                  / \________,'|
                 (((/  |       |
                 //    |       |
                //     |\      |
               //      | \     |
              //       |  \    |
             //        |   \   |

 Wow! It's impressive you're {user_age} years old!
 Although we respect all your experience, we do not recommend this game
 for people of your age. Try something a little less stressing ;) .\n
                    """)
                    endgame() #This takes us to the Fail Scenario
                    break

                else:          
                    context(f"""
 {guide_name}: Unbelievable, you're the right person, at the right time!
 Come with me!
                    """) 
                stage_1() #This takes us to Stage 1
                break
            else: 
                print("We don't think this is a realistic age to play this game, please try again ;)")
                age = input("<Insert here your age using numbers i.e 30\n > ")
                
        except ValueError:
            print("Please, remember to use integer numbers. Something like 18, 21, or 50")
            age = input("<Insert here your age using numbers i.e 30\n > ")

### Defining Stage 1: The Pills
def stage_1():
    """This function represents the introduction and initial context the user receives
    before having to make a choice. Red or Blue.
    Once done, it pops up the function pill_decision."""
          
    input("< Press Enter to continue > \n")
    print("_"*50)

    print(f"""
 {user_name}: Where are we? What is all this?""")
   
    #A bit of context for the user to understand what's going on.
    context(f"""
 {guide_name}: You have been living inside a simulation {user_name}. 
 Everything as you know it, including your {user_age} years, it's all part of a program.\n
 Today you'll have to decide if you want to face reality and fight
 for the survival of the human race ...\n
         """)
    
    input("< Press Enter to continue > \n")
    print("_"*50)
    
    context(f"""{user_name}: And if I don't want to ...?\n""")
    context(f"""
 {guide_name}: If you decide not to fight, you will wake up in your bed. 
 You won't remember any of this, and you'll continue with your "life".
 That is of course, if the rest of us manage to survive tonight ...\n
         """)
    input("< Press Enter to continue > \n")
    print("_"*50)
    context(f"""
 {guide_name}: It's time to make your choice.
 Here I have two pills. One red. One blue.\n""") #Explanation of the 1st decision to make.
    print("**********")
    context("If you take the RED pill, we'll start your training right away.\n")
    print("**********\n")
    print("**********")
    context("If you take the BLUE pill, it's the end of the story as I said before.\n")
    print("**********\n")
    
    context(f"{guide_name}: What is it going to be {user_name} ?")
    pill_decision() #Calls the next function
    
### Defining Stage 1.5: The Decision    
def pill_decision(): 
    """This function is used as the stage 1.5. 
    The user is prompted to decide if he/she is going to keep up with the challenge or not.
    Red Pill -> stage_2 / Blue Pill -> endgame.
    Here you will see nested conditionals and multiple error validations"""
    
    user_decision = input("Pick a number:\n1. Take the Red Pill\n2. Take the Blue Pill\n")
    user_decision = user_decision.lower() #formats strings in lower case to help the following validations
    
    #If user says 'yes', it will be requested a confirmation.
    if user_decision == '1' or user_decision == 'one' or user_decision == 'red':
        double_check = input("There is no coming back, are you sure?\n1. Yes\n2. No\n")
        double_check = double_check.lower()
        
        if double_check == '1' or double_check == 'yes': #This is the path to move forward
            context(f"{guide_name}: You did the right thing {user_name}. Follow me!")
            stage_2() #Moves to stage 2
                
        elif double_check == '2' or double_check == 'no': 
            context(f"{guide_name}: take 5 seconds to think about it")
            time.sleep(2)
            pill_decision() #returns to the beginning of this same function
        
        else:
            print("< Please, only use the number 1 or 2 >\n")
            pill_decision() #returns to the beginning of this same function
    
    #If user says 'no', it will end the game.        
    elif user_decision == '2' or user_decision == 'two' or user_decision == 'blue':
        context(f"""
 {guide_name}: I'm so dissapointed. 
 I really thought you were the one {user_name}...\n
         """)
        endgame() #Calls the fail function
        
    else:
        print("< Please, only use the number 1 or 2 >\n")
        pill_decision()        
        
### Defining Stage 2: The Place
def stage_2():
    """This function is used as the stage 2. 
    The idea is to have a bit more of user input -favorite place- in order to customize his/her experience.
    Once done, it pops up the function start_training."""
    
    time.sleep(1)
    input("< Press Enter to continue > \n")
    print("_"*50)
    context(f"{user_name}: Ok, how do we start?")
    context(f"""
 {guide_name}: You have been selected because your brain is unique! 
 No one has been able to complete the training 100%. Today you will be doing it!
 To help you a little bit, I will set the training arena just like a place you love.\n
         """)
    #asks for a place the user loves to personalize a couple of messages
    user_place = input("Write here your favorite place in the world:\n")  
    user_place = user_place.capitalize() #formats the string capitalizing the first letter
    context(f"{guide_name}: {user_place} it is!\n")
    time.sleep(1)
    print('****')
    time.sleep(0.5)
    print('***')
    time.sleep(0.5)
    print('**')
    context(f"{guide_name}: Done!\n\n")
    time.sleep(2)
    context(f"{user_name}: Wow, now this place looks exactly like {user_place}. Let's do it!\n")
    input("< Press Enter to continue > \n")
    print("_"*50)

    start_training() #Calls the next stage
    
### Defining Stage 3: The Training
def start_training():
    """This function is used as the stage 3. 
    Using a loop, the % of training reaches 50%. This is expected and is used as preparation.
    However, the user will face a control question in order to unlock the other 50% of the trainig.
    In a future version of the game, this and the other stages can be used as 'checkpoints'."""
    
    print('*** Training Started ***')
    time.sleep(0.5)
    comp_training = 0
    
    #loop to create an animation effect from 0 to 50%
    while comp_training < 50:
        comp_training += 10
        print(f"""*** {comp_training}% completed ***""")
        time.sleep(1)
    
    context(f"{guide_name}: Great!! the first half is done!\n\n")
    context(f"{user_name}: Thank you, I can't wait to complete the other 50%!\n\n")
    context(f"""
 {guide_name}: Now it starts to get a bit trickier.\n
 I hope you feel specially lucky today!\n
         """)
    input("< Press Enter to continue > \n")
    print("_"*50)
    context(f"{user_name}: Actually, I do!\n")
    context(f"""
 {guide_name}: Awesome, let's keep going.\n
 The training system requires you to guess correctly the following question.
 Be aware that you will only have 2 chances!
    """)
    input("< Press Enter to receive the question > \n") 
    print("_"*50)
    
    #First guess begins
    picks = 2 #Variable for the number of chances before failing the game.
    random_number = randint(1,3) #Random number between 1 and 3 inclusive. This will be the correct answer.
    expected_number = str(random_number) #Converts the random number to string to help with validations
    user_number = input("How many weak spots do the machines have?\na.1\nb.2\nc.3 \n" )
    
    #general loop to control the user's lives and choices
    while picks > 0: 
            if user_number == '1' or user_number == '2' or user_number =='3':
                if user_number == expected_number: # User guessed and goes to the next stage!
                    print("CORRECT!") 
                    picks = 0
                    continue_training() #Calls next stage
                    break #Stops the loop
                else:
                    picks -= 1 # User loses 1 chance
                    if picks > 0: #If user has 1 more chance
                        print(f"Incorrect!\n You have {picks} chance left\n")
                        user_number = input("How many weak spots do the machines have?\na.1\nb.2\nc.3 \n" )
                    elif picks == 0: #If user lost his/her final life
                        print(f"Incorrect! The right answer was {expected_number}\n")
                        endgame() #Calls fail function
                        break
            else: 
                print("For your guess, please input the number 1, 2 or 3\n")
                user_number = input("How many weak spots do the machines have?\na.1\nb.2\nc.3 \n" )

### Defining Stage 3.5: The 2nd Part of the Training
def continue_training():
    """This function is used as the stage 3.5. 
    Using a loop, the % of training reaches 80%, until it's interrupted by the machine.
    In a future version of the game, this and the other stages can be used as 'checkpoints'."""
    
    context(f"""
 {guide_name}: YES!! \n
 Definitely you have something special {user_name}!
 Let's continue with your training.\n
    """)
    input("< Press Enter to continue > \n")
    print("_"*50)
    
    print('*** Training Resumed ***') #Training starts again. Now from 50%
    time.sleep(1)
    comp_training = 50
    
    #loop to create animation of the training
    while comp_training < 80:
            comp_training += 10 
            print(f"""*** {comp_training}% completed ***""")
            time.sleep(0.5)
    print ("WARNING! WARNING! WARNING! System Stopped ... ") #prepares user for action
    input("< Press Enter to continue > \n")
    print("_"*50)
    context(f"""
 {guide_name}: Oh no! \n
 We have an emergency {user_name}!
 One Machine found our secret cave and it's coming to attack us!\n\n
        """)
    context(f"{user_name}: No way! What do we do?\n\n")
    time.sleep(1)
    context(f"{guide_name}: WE FIGHT!\n") #Motivational Answer!
    input("< Press Enter to continue > \n")
    stage_4() #Calls the most exciting stage
    
### Defining Stage 4: The Fight
def stage_4():
    """This function is used as the stage 4. 
    Using nested conditionals inside a main loop, the user will face the final level of the game.
    The dificulty of this level is really high due to the changing random numbers.
    For testing purposes, you as a visitor programmer can go to the variable 'weak_spot'
    and define it with a single integer value from 1 to 3. -look for the line 3466 or the 'TEST HERE' comment.
    In a future version of the game, this and the other stages can be used as 'checkpoints'."""
    
    print("_"*50)
    
    #Explains what is need to be done
    context(f"""
 {guide_name}: Here it comes! \n
 {user_name}, listen to me, only you can save us! The rest of the crew will prepare
 the final electromagnetic pulse a.k.a -EMP-. However, in order to be effective, 
 first you have to cause critical damage to that machine!\n  
         """)
    input("< Press Enter to continue > \n")
    
    #Explains how process works
    context(f"""
 {guide_name}: Now! Attack 1 of the 3 possible spots of the machine.
 Only 1 is correct and it changes everytime.
 If you succeed, the machine will lost 30% of its life. 
 Nevertheless, with every attack you do (either right or wrong), you'll lose 10% of yours.
 """)
    machine_life = 100 #Machine_life starts at 100
    user_life = 50 # Since the user couldn't finish the training, his/her life starts at 50 // 
    print("Machine Life: "+"*"* (machine_life))
    print("Your Life: "+"*"* (user_life))
    
    while machine_life > 0 and user_life > 0: #while both are alive, keep going!
        print("_"*50)
        
        #Variable that takes, the value of user's decision
        move = input("Which spot are you going to attack?:\na.1\nb.2\nc.3\n")
        
        #Simulates a graphic for the "lifebars"
        print("Machine Life: "+"*"* (machine_life))
        print("Your Life: "+"*"* (user_life))
        weak_spot = 3 #randint(1,3) #<- TEST HERE! Random number changes every iteration to increase the difficulty
        
        try: #In this case, the random stays as int, and we'll convert the move variable from str to int
            attack = int(move)
            if attack >= 1 and attack <=3:
                if attack == weak_spot and machine_life == 40 and user_life > 10: #After this, the machine will be gone!
                    machine_life -= 30  
                    user_life -= 10
                    print("Machine Life: "+"*"* (machine_life))
                    print("Your Life: "+"*"* (user_life))
                    print(f"""
     >> Good job! Now the machine has {machine_life}% of life remaining.
     You also suffered and your life is at {user_life}%.
                """)
                    machine_life = 0 # The real machine life was 10, which is enough to be affected by the EMP. 
                    context(f"""
     >> {user_name}: The machine is really damaged!
     It's the right moment to activate the EMP.
     Hurry {guide_name} it's now or never!
        """)
                    input("< Press any key to continue > \n")
                    activate_emp() # This is 1 of 2 options to win the game!
                    break

                elif attack == weak_spot and machine_life > 40 and user_life > 10: #Regular positive attack.
                    machine_life -= 30
                    user_life -= 10
                    print("Machine Life: "+"*"* (machine_life))
                    print("Your Life: "+"*"* (user_life))
                    print(f"""
     >> Good job! Now the machine has {machine_life}% of life remaining.
     You also suffered and your life is at {user_life}%.
                """)

                elif attack == weak_spot and machine_life <= 40 and user_life == 10: #After this, the machine will be gone!
                    machine_life -= 30
                    user_life -= 10
                    print("Machine Life: "+"*"* (machine_life))
                    print("Your Life: "+"*"* (user_life))
                    print(f"""
     >> Good job! Now the machine has {machine_life}% of life remaining.
                """)
                    context(f"{user_name}: {guide_name}! Please come here ... ")
                    time.sleep(1)
                    context(f"{guide_name}: Here I am, you're doing great {user_name}!")
                    time.sleep(1)
                    context(f"""
     >> {user_name}: Listen ... unfortunately, I won't be able to survive with these wounds.
     The good thing is the machine is really damaged! It's the right moment to activate the EMP.
     Hurry {guide_name}, it's now or never!
     """)      
                    activate_emp() # This is the 2nd option to win the game! User defeats the machine but dies as well 
                    break          # Fun Fact: On Matrix 3, Neo the protagonist, sacrifices himself to save the resistance!

                elif attack == weak_spot and machine_life > 40 and user_life == 10: #After this, the user dies!
                    machine_life -= 30
                    
                    context(f"""
     >> Although you hit it, the machine still has {machine_life}% of life remaining.
     You just lost your last {user_life}%.
                """)
                    user_life = 0 #With this I ensure the while finishes
                    print("Machine Life: "+"*"* (machine_life))
                    print("Your Life: "+"*"* (user_life))
                    endgame() #Calls the fail function
                    break

                else:
                    user_life -= 10 #Regular wrong attack. 
                    print("Machine Life: "+"*"* (machine_life))
                    print("Your Life: "+"*"* (user_life))
                    print(f"""
     >> Wrong choice! You got hit and your life is at {user_life}%.
                """)
            else:
                print("For your guess, please input the number 1, 2 or 3\n")
                move = input("Choose wich spot are you going to attack:\na.1\nb.2\nc.3\n")
        except ValueError:
            print("For your guess, please input the number 1, 2 or 3\n")
        
### Defining Winning Fuction: EMP    
def activate_emp():
    """This function is used as a winning method. 
       Once done, it pops up the function play_again."""
    
    guide_name = '* Morpheus' #Just for "design" purposes
    print("_"*50)
    
    context(f"{guide_name}: All set!\n")
    time.sleep(0.5)
    print("3\n")
    time.sleep(0.5)
    print("2\n")
    time.sleep(0.5)
    print("1\n")
    time.sleep(0.5)
    context('*** EMP ACTIVATED ***\n') #Our favorite tool from now on!
    time.sleep(1)
    print ("*" * 50)        
    print ("*" * 50)     
    print ("*" * 50)        
    print ("*" * 50) 
    input("< Press any key to continue > \n")
    print("_"*50)
    context(f"""
 {guide_name}: THANK YOU {user_name}. You have saved the entire human race\n
 I'll make sure everyone knows you are a legend!
 """)
    context("YOU WIN!!!\n") #VICTORY!!
    print ("*" * 50)        
    print ("*" * 50)     
    print ("*" * 50)        
    print ("*" * 50)
    input("< Press any key to continue > \n")
    play_again() #Calls the replay function

### Defining Failing Fuction: Endgame ---> Yes, I love The Avengers!
def endgame():
    """This function is used as a failure method. 
    Once done, it pops up the function play_again.""" 
    
    context("Unfortunately, your path ends here.\n") #Previous message before asking if user wants to play again
    input("< Press any key to continue >\n ")
       
    print ("*" * 50)
    print ("*" * 50)
    print ("*"*19+" GAME OVER "+"*"*20) # A classic!
    print ("*" * 50)        
    print ("*" * 50) 
    
    play_again() #Calls the replay function

### Defining Replay Fuction    
def play_again():
    """This function offers the chance to start over once you have failed.
    Keep in mind that you'll have to start from the beginning."""
    
    context("Thank you for playing THE HULTRIX\n") #Thank you for reading even the comments!      
    replay = input("Do you want to play again? (Y/N)\n") #Willing to play again?
    replay = replay.lower()
    if replay == 'y':
        
        print("***")
        print("**")
        print("*\n")
        time.sleep(1)
        start_game() #Calls the initial stage! Let's try the different options.
          
    elif replay == 'n':
        time.sleep(2)
        context(f"Until the next time {user_name}.")
        exit(0) #Shuts down the execution - Valid for this assignment.
    
    else:
        print("Please make your choice using letters 'Y' or 'N'.")
        play_again() #Validation
    
          

        
###GAME STARTER###
start_game() #Calls the initial stage! Let's try the different options.       
```
